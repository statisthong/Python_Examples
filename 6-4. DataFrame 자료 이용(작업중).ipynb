```python
dataL = [['n0', 'a0', 'b0', 'c0'], ['n0', 'a1', 'b1', 'c1'], ['n0', 'a2', 'b2', 'c2'], ['n1', 'a3', 'b3', 'c3']] 
```

```python
dataL
```

```python
import pandas as pd
dfL = pd.DataFrame(dataL)
dfL
```

```python
dfL.columns = ["N", "A", "B", "C"]; dfL
```

```python
dataR = [['n0', 'a0', 'b0', 'd0'], ['n0', 'a1', 'b1', 'd1'], ['n1', 'a2', 'b2', 'd2'], ['n1', 'a3', 'b3', 'd3']] 
```

```python
dataR
```

```python
import pandas as pd
dfR = pd.DataFrame(dataR)
dfR
```

```python
dfR.columns = ["N", "A", "B", "D"]; dfR
```

```python
pd.merge(dfL, dfR, on='N')
```

```python
df = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']});df
```

```python
other = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']}); other
```

```python
df.join(other, how='inner', lsuffix='_caller', rsuffix='_other') # how : left(기본값), right, inner lsuffix 왼쪽 접두어, rsuffix 오른쪽 접미어
```

```python
dfL.join(dfR, on='N')
```

```python
df.set_index('key').join(other.set_index('key'), how='inner')
```

```python
df.join(other.set_index('key'), how='inner', on='key')
```
