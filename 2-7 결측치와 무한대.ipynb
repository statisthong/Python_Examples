* 결측치

- numpy, pandas 모듈 내에서 지원 됨
- R의 NA와 같은 기증은 지원되지 않고 NaN, Inf와 동일한 기능 지원
- Not a Number : nan, NaN, NAN
- 무한대 : inf, Inf, infty, Infinity 중 하나를 사용

```python
import numpy as np
```

```python
height = np.array([45, 75, 60, 51, np.nan])
```

```python
height = np.array([45, 75, 60, 51, np.nan])
```

```python
height
```

```python
np.mean(height) # nan이 있으므로 평균 계산이 안됨
```

```python
np.nanmean(height) # nanmean : NaN을 제외한 평균 계산을 해주는 함수
```

```python
x = np.array([1, -1, np.infty, -np.inf])
```

```python
x
```

```python
print(x/0)
```

```python
print(x/np.Infinity) # R에서는 0, 0, NAN, NAN 가 됨
```
