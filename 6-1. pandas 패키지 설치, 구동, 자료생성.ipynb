```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

* Series : 각 원소가 이름을 갖는 1차원 혼합 자료형 배열
* DataFrame : 행이름과 열이름이 있는 크기 변경이 가능한 테이블 구조의 배열

```python
s = Series([10, 20, 30], index=['a', 'b', 'c']); s
```

```python
df = DataFrame(s, columns=["A"]); df
```
