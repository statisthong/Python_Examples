* 파이썬의 벡터 연산(R과 유사한 형태)

```python
xcol = np.array([[1],[2],[3]]); xcol # 열벡터
```

```python
print("xrow \n", xrow, "\n xcol: \n", xcol) # \n : 줄바꿈 특수문자
```

* 행렬과 벡터의 항목 추가, 삭제 
* 파이썬의 첨자 연산 사용보다 numpy 패키지의 delete, append, insert 등의 함수를 사용하여 수행하는 것이 편리



```python
import numpy as np
a = np.array(range(1,11,2))
```

```python
a
```

```python
a1 = np.delete(a,2) # ndarray a에서 3번째 항목(인덱스 2) 삭제, 
```

```python
a1
```

```python
idx = [0,1,3] # 인덱스 0,1,3
a2 = np.delete(a, idx) # 한번에 여러게 삭제
a2
```

```python
b1 = 11
a3 = np.append(a, b1) # ndarray a에 b1 추가
```

```python
a3
```

```python
b2 = 2
a4 = np.insert(a, 1, b2) # ndarray a 인덱스 2번째 위치에 b2를 삽입
a4
```

```python
b3 = [1,2,3]
a5 = np.insert(a, 2, b3) # ndarray a 인덱스 3번째 위치에 b3를 삽입
a5
```

* 행렬에서의 추가, 삽입, 삭제

```python
mx1 = np.array(range(1,10)).reshape(3,3)
mx1
```

```python
mx2 = np.append(mx1, np.array([10,20,30]).reshape(1,3), axis=0) # 행벡터 추가, axis=0 행
mx3 = np.append(mx1, np.array([10,20,30]).reshape(3,1), axis=1) # 열벡터 추가, axis=0 열
```

```python
mx2
```

```python
mx3
```

* 행, 열 삭제

```python
mx4 = np.delete(mx1, 0, axis=0) # mx1 행렬에서 첫째 행 삭제
mx5 = np.delete(mx1, [0,1], axis=1) # xm1 행렬에서 1,2 열 삭제
```

```python
mx4
```

```python
mx5
```

* 행,열 삽입

```python
mx6 = np.insert(mx1, 1, np.array([10,20,30]), axis=0) # mx1 행렬의 두번째 위치에 행으로 삽입
mx7 = np.insert(mx1, 1, np.array(range(10,100,10)).reshape(3,3), axis=1) # mx1 행렬의 두번째 위치에 열로 삽입
```

```python
mx6
```

```python
mx7
```
