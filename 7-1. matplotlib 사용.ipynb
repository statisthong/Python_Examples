* plt.ioff() : 이후의 입력을 비대화형 방식으로 하여 그림을 구성함
* plt.show() : 비대화형 방식으로 앞서 입력하여 구성된 그림을 그림

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
%matplotlib inline
```

```python
x = np.random.randn(100)
y = np.random.randn(100)
```

```python
plt.plot(y, 'b:')
```

```python
plt.title('Green Solid Line')
```

```python
plt.title('Blue Dotted Line')
```

```python
plt.ioff()
```

```python
plt.figure()
```

```python
plt.plot(y, 'm-') # g: green - 선모양, b: blue, r: red, c: cyan, m: magenta, y: yellow, k: black, w: white
```

```python
plt.title('Green Solid Line')
```

```python
plt.show()
```

```python
plt.ion()
```

```python
plt.figure()
```

```python
h = plt.plot(y, 'g-')
```

```python
plt.getp(h)
```

```python
plt.setp(h, 'color', 'red')
```

```python
plt.figure()
```

```python
plt.plot(y, alpha=0.5, ls='-', lw=3, marker='o', mec='red', mew=2, mfc="blue", ms=10) # lw(linewidth) : 선굵기, marker : 마커 종류, mew(markeredgewidth) : 마커선 굵기, ms(marker size): 마커 크기, mfc(markerfacecolor) : 마커 내부 색깔
```
