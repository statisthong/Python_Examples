* 문자열 비교, 찾기, 교체

```python
mystr = 'abcdeabcab'
```

```python
n0 = mystr.count('ab')
```

```python
n0 # ab가 3번 총 나옴
```

```python
n1 = mystr.find('ab')
```

```python
n1 # ab가 처음 나오는 위치 : 0번째 인덱스 
```

```python
n2 = mystr.rfind('ab')
```

```python
n2 # ab를 오른쪽에서부터 찾기, 인덱스는 8(0부터 시작함으로 9번째 임)
```

```python
newstr1 = mystr.replace('ab', 'AB') # ab 를 AB 로 모두 교체
```

```python
newstr 
```

```python
newstr2 = mystr.replace('ab', 'AB', 2) # ab 를 AB로 2번 교체
```

```python
newstr2
```
