* 상위클래스 없이 클래스를 새로 구성하는 방법

```python
class Student:
  def __init__(self, first, last): # 생성자, Student 클래스의 객체를 잡는 함수로 학생의 성(last)과 이름(first)을 받아 객체의 속성 self.first와 self.last로 설정
    self.first = first
    self.last = last
  def capital_first(self): # 이름을 대문자로 변환
    return self.first.upper()
  def capital_last(self): # 성을 대문자로 변환
    return self.last.upper()
```

```python
aa = Student('John', 'Doe')
bb = Student('Jane', 'Doe')
aa.first, aa.last # aa.first ==> self에 aa가 대응됨
```

```python
bb.first, bb.last # bb.first ==> self에 bb가 대응됨
```

```python
aa.capital_first()
```

```python
bb.capital_first()
```

```python
aa.capital_last()
```

```python
bb.capital_last()
```

```python
aa.__class__
```

```python
bb.__class__
```

* 상위 클래스를 잡고 이를 변형하여 하위 클래스를 구성하는 방법

* 상위 클래스인 Student 클래스에 성적(Score) 속성을 추가하고 80점 미만인 경우 B 학점을 주고, 80점 이상인 경우 A 학점을 주는 grade 메소드를 추가하여 StudentGrade라는 하위 클래스를 구성하는 방법

* 클래스를 정의할 때 상위 클래스 이름은 괄호 안에 포함하여 StudentGrade(Student)와 같이 클래스를 구성 함

```python
class StudentGrade(Student):
  def __init__(self, first, last, score): # self.first, self.last, self.score
    super().__init__(first, last) # self.first = first, self.last = last
    self.score = score
  def grade(self):
    if self.score < 80:
      grade = 'B'
    else:
      grade = 'A'
    return grade
```

```python
aa = StudentGrade('John', 'Doe', 75); aa
```

```python
bb = StudentGrade('Jane', 'Doe', 95); bb
```

```python
aa.first, aa.last
```

```python
bb.first, bb.last
```

```python
aa.score, aa.grade()
```

```python
bb.score, bb.grade()
```

```python
aa.capital_first() # 상위 클래스 Student에서 정의된 것을 그대로 상속되어 StudentGrade 의 속성으로 사용
```

```python
bb.capital_first()
```
