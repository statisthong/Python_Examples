```python
import re
```

```python
mystr = re.sub('\\[', 'A', '[12.5+13-1*5$') # 특수문자 ] 를 A로 바꿈, \ 자체가 특수문자를 의미, \를 표시하기 위해 \\로 두번 사용함
```

```python
mystr
```

```python
print(mystr)
```

```python
mystr = re.sub('[a-z]', '-', 'bcde0123ABC') # a부터-z까지의 영어소문자를 모두 -로 바꿈
```

```python
print(mystr)
```

```python
mystr = re.sub('[a-z]', '-', 'bcde0123ABC',2) # a부터-z까지의 영어소문자를 2번만 -로 바꿈
```

```python
print(mystr)
```

```python
l_case = re.search('[a-z]', 'bcde0123ABC') # 문자열 'bcde0123ABC' 에서 소문자 a부터 z까지 하나라도 있으면 True, 아니면 False
```

```python
print(l_case)
```

```python
if l_case: print('a-z exists') 
else: print('No a-z')
```

```python
l_case = re.search('[k-z]', 'bcde0123ABC') # 문자열 'bcde0123ABC' 에서 소문자 k부터 z까지 하나라도 있으면 True, 아니면 False
```

```python
if l_case: print('k-z exists') 
else: print('No k-z')
```

```python
mystr3 = [re.sub('colou?r', 'COLOR', word) for word in ['color', 'coloor', 'colooor', 'colour', 'colouur']] # ? colou?r : u 자리에 아무것도 없거나 u가 1개 있는 "color"과 "colour"와 일치함
```

```python
mystr3
```

```python
mystr4 = [re.sub('colou*r', 'COLOR', word) for word in ['color', 'coloor', 'colooor', 'colour', 'colouur']] # * ab*c : "ac", "abc", "abbc", "abbbc" 과 일치함
```

```python
mystr4
```

```python
mystr5 = [re.sub('colou+r', 'COLOR', word) for word in ['color', 'coloor', 'colooor', 'colour', 'colouur']] # + ab+c : "abc", "abbc", "abbbc" 과 일치함. "ac"와는 일치하지 않음
```

```python
mystr5
```

```python
mystr6 = [re.sub('colo{3}r', 'COLOR', word) for word in ['color', 'coloor', 'colooor', 'colour', 'colouur']] # {} a{3} : aaa, a 4번 반복함
```

```python
mystr6
```

```python
mystr7 = [re.sub('colo{1,3}r', 'COLOR', word) for word in ['color', 'coloor', 'colooor', 'colour', 'colouur']] # {min, max} : 최소 min 번 부터 최대 max 번까지 반복함
```

```python
mystr7
```

```python
mystr8 = [re.sub('colo{2,}r', 'COLOR', word) for word in ['color', 'coloor', 'colooor', 'colour', 'colouur']] # {min,} : min 이상 번을 반복함
```

```python
mystr8
```

```python
[re.sub('a.b', 'A.B', word) for word in ['ab', 'abb', 'abbb', 'axb', 'axxxxb']] # . 마침표는 1개의 임으의 문자 (wild card)
```

```python
[re.sub('a.*b', 'A.B', word) for word in ['ab', 'abb', 'abbb', 'axb', 'axxxxb']] # * 표는 임의의 길이의 문자열 (wild card)
```

```python
[re.sub('^1', '-', word) for word in ['12.5', 'ab', 'abc', '']] # ^ : 문자열의 시작 또는 부정(not), ^a : 문자열의 처음에 있는 a
```

```python
[re.sub('^a', '-', word) for word in ['12.5', 'ab', 'abc', '']] # ^ : ^a : 문자열의 처음에 있는 a
```

```python
[re.sub('[^0-9]', '-', word) for word in ['12.5', 'ab', 'abc', '']] # [^0-9] : 0에서 9사이의 숫자가 아닌 다른 모든 문자
```

```python
[re.sub('b$', '-', word) for word in ['12.5', 'ab', 'abc', '']] #b$ : b로 끝나는 문자열
```

```python
[re.sub('b\^$', '-', word) for word in ['12.5', 'ab', 'abc', '']] # ^$ : 빈(empty) 문자열
```
