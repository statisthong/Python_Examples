* 리스트 : 원소가 어떤 형태이든 상관없이 저장 가능, 행렬과 벡터라는 수학적 의미보다는 여러개를 하나의 이름으로 저장하기 위한 용도



```python
a = [2, 3, 'c']
x = [1, 2, 'a', a]
print(x)
```

```python
x[0] = 0; x[0] # array와 같은 방법으로 x[0] 값 위치 설정
```

```python
x[3]
```

```python
x2 = [[1,2,3], [4,5,6], [7,8,9]] # 행렬 형태의 2차원 리스트
x2
```

```python
len(x2)
```

```python
x2[0][2] # 0행 2열
```

```python
x2[2] # 2차원 리스트의 경우 특정한 행 전체를 얻으려면 행번호만 []에 입력
```

```python
[col[2] for col in x2] # 특정한 열을 얻기 위해 for문 사용
```

```python
x2.reshape(3.3) # array가 아니라 리스트이으로 reshape를 사용할 수 없음
```

```python
x2.__class__
```

```python
type(x2)
```

* 튜플(tuple) : 리스트와 유사, 원소값 설정에 () 사용, 최초 설정된 이후 수정이 안됨

```python
tpl = (1,2,3)
```

```python
type(tpl)
```

```python
tpl[2]
```

```python
tpl[2] = 3
```

* 딕셔너리(Dictionary) : 인덱스가 0부터 시작하는 정수가 아니라, 사전처럼 특정한 이름을 갖도록 함

```python
country_code = {'korea':82, 'us':1, 'china':86}
```

```python
country_code
```

```python
print(country_code)
```

```python
print(len(country_code))
```

```python
print(country_code['korea'])
```

```python
print(country_code.keys()) # 키(Key)만 출력
```

```python
print(country_code.values()) # 해당 값만 출력
```

```python
print(country_code.items()) # 키와 값의 짝으로 출력
```

* 리스트와 배열 복사시 주의할 점

```python
xm = [[1,2,3], [4,5,6], [7,8,9]];xm
```

```python
xm2 = xm # xm 과 xm2는 같은 주소를 사용하여 저장됨. xm이나 xm2 중 하나만 값을 바꿔도 둘 다 바뀌게 됨
```

```python
xm2 = xm.copy()  # copy 사용하여 xm2를 만들어도 결과는 같음
```

```python
xm[0][0] = 10;xm2
```

```python
ym = [[1,2,3], [4,5,6], [7,8,9]]
```

```python
import copy as cp
ym2 = cp.deepcopy(ym)
```

```python
ym[0][0] = 10
```

```python
ym2[0][0] #ym[0][0] 의 값이 바뀔때 ym2[0][0]의 값은 바뀌지 않음
```
